## 2. Theory / Background

**LSTM** is a powerful Recurrent Neural Network architecture designed to handle long-term dependencies in sequential data. It is widely used in time series forecasting because:

- It uses memory cells and three gates (forget, input, output) to regulate information flow.
- It effectively captures trends, seasonality, and long-range patterns.
- It solves the vanishing gradient problem of standard RNNs.

**Key Components:**
- **LSTM Layer**: Learns temporal dependencies from sequential input.
- **Flatten**: Converts LSTM output into a 1D vector.
- **Dense(1)**: Produces the final predicted value.

# 4. LSTM Model Architecture

```python
def create_lstm():
    input_data = Input(shape=(time_steps, num_features))
    lstm_layer1 = LSTM(8, return_sequences=True)(input_data)
    lstm_layer2 = LSTM(20)(lstm_layer1)
    x = Flatten()(lstm_layer2)
    output_data = Dense(1)(x)
    model = Model(input_data, output_data)
    return model
```
